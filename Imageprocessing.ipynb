```python
from google.colab.patches import cv2_imshow
import cv2
from skimage import filters
import numpy as np
from matplotlib import pyplot as plt
```

```python
# Reading a image
image_path = '/content/cat.jpg'
image = cv2.imread(image_path)
```

```python
# Display the original image
cv2_imshow( image)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
# Linear Filter (Blur)
linear_filtered_image = cv2.blur(image, (5, 5))
```

```python
# Display the linear filtered image
cv2_imshow( linear_filtered_image)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
# Non-linear Filter (Median)
non_linear_filtered_image = cv2.medianBlur(image, 5)
```

```python
# Display the non-linear filtered image
cv2_imshow(non_linear_filtered_image)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
# Non-linear Filter (Max or Dilation)
max_filtered_image = cv2.dilate(image, None, iterations=2)
```

```python
# Display the max filtered image
cv2_imshow(max_filtered_image)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
# Non-linear Filter (Median)
non_linear_filtered_image = cv2.medianBlur(image, 5)
```

```python
# Display the non-linear filtered image
cv2_imshow( non_linear_filtered_image)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
#sobel filter for edge detection
sobel_filtered_image_x = cv2.Sobel(image, cv2.CV_64F, 1, 0, ksize=5)
sobel_filtered_image_y = cv2.Sobel(image, cv2.CV_64F, 0, 1, ksize=5)
sobel_filtered_image = np.sqrt(sobel_filtered_image_x**2 + sobel_filtered_image_y**2)
```

```python
# Normalize the Sobel filtered image to be in the range [0, 255]
sobel_filtered_image = cv2.normalize(sobel_filtered_image, None, 0, 255, cv2.NORM_MINMAX)
```

```python
# Convert to uint8 for display
sobel_filtered_image = np.uint8(sobel_filtered_image)
```

```python
# Display the Sobel filtered image
cv2_imshow(sobel_filtered_image)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
# using Canny Filter for detecting diagnol edges
canny_filtered_image = cv2.Canny(image, 50, 150)
```

```python
# Displaying the Canny filtered image
cv2_imshow(canny_filtered_image)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
# second deriative filters
# laplacian filter
laplacian_filtered_image = cv2.Laplacian(image, cv2.CV_64F)
```

```python
# Normalize the Laplacian filtered image to be in the range [0, 255]
laplacian_filtered_image = cv2.normalize(laplacian_filtered_image, None, 0, 255, cv2.NORM_MINMAX)
```

```python
# Convert to uint8 for display
laplacian_filtered_image = np.uint8(laplacian_filtered_image)
```

```python
# Display the Laplacian filtered image
cv2_imshow(laplacian_filtered_image)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
# Laplacian of Gaussian (LoG) Filter
log_filtered_image = cv2.GaussianBlur(image, (5, 5), 0)  # Gaussian smoothing
log_filtered_image = cv2.Laplacian(log_filtered_image, cv2.CV_64F)
```

```python
# Normalize the LoG filtered image to be in the range [0, 255]
log_filtered_image = cv2.normalize(log_filtered_image, None, 0, 255, cv2.NORM_MINMAX)
```

```python
# Convert to uint8 for display
log_filtered_image = np.uint8(log_filtered_image)
```

```python
# Display the LoG filtered image
cv2_imshow( log_filtered_image)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
# shape detection filters
#frangi
frangi_filtered_image = filters.frangi(image)
```

```python
# Display the Frangi filtered image
cv2_imshow( frangi_filtered_image)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
